This script is used to read the raw data and generate `education.csv`.

```python
import pandas as pd
import numpy as np

from constants import EDUCATION_DIMENSION_CSV
```

```python
df = pd.read_csv('../raw_data/education.csv')
df.replace("..", np.nan, inplace=True)
```

```python
columns = df.columns.tolist()
# fix column years format
for i in range(4, len(columns)):
	columns[i] = columns[i][:4]
df.columns = columns
df.head(6)
```

```python
# Replace missing value with the mean value for each country and attribute 
df[columns[4:]] = df[columns[4:]].T.fillna(df.mean(axis=1)).T
df.head(6)
```

```python
attributes_dic = {
    "public_spending_percent_of_gdp": "SE.XPD.TOTL.GD.ZS",
    "literacy_rate_adult_total": "SE.ADT.LITR.ZS",
    "literacy_rate_adult_male": "SE.ADT.LITR.MA.ZS",
    "literacy_rate_adult_female": "SE.ADT.LITR.FE.ZS",
    "primary_school_enrollment_percent_of_gross": "SE.PRM.ENRR",
    "primary_school_enrollment_percent_of_net": "SE.PRM.NENR",
    "primary_completion_rate": "SE.PRM.CMPT.ZS",
    "primary_completion_male": "SE.PRM.CMPT.MA.ZS",
    "primary_completion_female": "SE.PRM.CMPT.FE.ZS",
    "secondary_school_enrollment_percent_of_gross": "SE.SEC.ENRR",
    "secondary_school_enrollment_percent_of_net": "SE.SEC.NENR",
    "tertiary_school_enrollment_percent_of_gross": "SE.TER.ENRR",
}
```

```python
def transpose_attributes(dataframe, attribute):
	dataframe.drop(columns=["Series Name", "Series Code", "Country Name"], inplace=True)
	columns = dataframe.columns.tolist()
	return dataframe.melt(id_vars="Country Code", value_vars=columns[1:], var_name="Year", value_name=attribute)
```

```python
# Transpose Attributes
attributes_df_dic = {}
for key, value in attributes_dic.items(): 
	filt = (df["Series Code"] == value)
	attribute_df = df.loc[filt].copy()
	attributes_df_dic[key] = transpose_attributes(attribute_df, key)
```

```python
# Merge attributes together
attributes_lst_dic = list(attributes_df_dic.values())
education_df = attributes_lst_dic[0] 
for i in range(1, len(attributes_lst_dic)):
	education_df = pd.merge(education_df, attributes_lst_dic[i], how="left", left_on=["Country Code", "Year"], right_on=["Country Code", "Year"])
education_df.rename_axis("education_key", inplace=True)
education_df.head(30)
```

```python
# Output Education Dimension
education_df.to_csv(EDUCATION_DIMENSION_CSV)
```
