```python
import pandas as pd
import numpy as np

from constants import POLITICAL_EVENT_DIMENSION_CSV
```

```python
df = pd.read_csv('../raw_data/political_events.csv')
```

```python
countries = ["Canada", "USA", "Mexico", "India", "Indonesia", "Iran", "Egypt", "Kenya", "Nigeria"]
year = 2005
filter = (df["countryname"].isin(countries)) & (df["year"] >= year)
political_event_df = df.loc[filter].copy()
political_event_df.reset_index(inplace=True)
political_event_df.rename_axis("political_event_key", inplace=True)
political_event_df.head(8)
```

```python
columns_rename_dic = {
	"ifs": "Country Code",
	"year": "Year",
}
columns_to_drop = ["index", "countryname"]

political_event_df.rename(columns=columns_rename_dic, inplace=True)
political_event_df.drop(columns=columns_to_drop, inplace=True)
political_event_df.head(14)
```

```python
political_event_df.to_csv(POLITICAL_EVENT_DIMENSION_CSV)
```
